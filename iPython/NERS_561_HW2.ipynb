# Native Data Manipulation and Plotting

```python
f851tab = [1.95,0.810, 0.684, 0.512, 0.350, 0.210, 0.111, 0.061, 0.0220]
f402tab = [2.38,0.584, 0.455, 0.290, 0.147, 0.046, 0.018, 0.0043, -0.0004]
f347tab = [8.1,0.0092, 0.00015, 0, 0, 0, 0, 0, 0]
```

```python
fig = plt.figure(figsize=(7,4.326))
ax = fig.add_subplot(111)
plt.title(r"");
plt.xlabel(r"$\mu_{\textrm{cm}}$");
plt.ylabel(r"$f\left(\mu_{\textrm{cm}}\right)$");

x = np.linspace(-1, 1, 201)

plt.ylim([-1,4])

plt.plot(x, np.polynomial.legendre.legval(x, f851tab).T/f851tab[0],'k-',label="8.51 MeV")
plt.plot(x, np.polynomial.legendre.legval(x, f402tab).T/f402tab[0],'k--',label="4.02 MeV")
plt.plot(x, np.polynomial.legendre.legval(x, f347tab).T/f347tab[0],'k:',label="34.7 keV")

plt.legend(loc=2)

#plt.show();
plt.savefig('NERS_561_HW2_Plot_1959_Data.pdf',bbox_inches='tight')
```
