# Read JANIS Data

```python
# Identify JANIS data file to read.
inpath = '/Users/jkulesza/Dropbox/CLASSES/2015_01_Winter/NERS_561/HW3/'
infile = 'JEFF-32_H-1_MT_1_2_102_Cross_Section_Orig_Data_100_per_Decade.csv'

# Read JANIS CSV file in (skipping first 3 (header) rows).
xsfile=np.loadtxt(open(inpath+infile,"rb"), delimiter=",", skiprows=3)

# Transform energy to lethargy (reset first column as Log(E0 / E)).
xsfile[:,0] = np.log(E0 / xsfile[:,0])

# Convert microscopic data to macroscopic data.
xsfile[:,1] = N_H * xsfile[:,1]
xsfile[:,2] = N_H * xsfile[:,2]
xsfile[:,3] = N_H * xsfile[:,3]

# Create interpolated functions to read cross-section data at arbitrary 
# lethargies.
sigma_s = interp1d(xsfile[:,0], xsfile[:,1])
sigma_a = interp1d(xsfile[:,0], xsfile[:,2])
sigma_t = interp1d(xsfile[:,0], xsfile[:,3])
```

# Initialize Numeric Solver

```python
def inner_integral(u,phi,jmax):
    integral = 0
    for j in range(1,jmax):
        deltauj = u[j] - u[j-1]
        integral = integral + (deltauj * sigma_s(u[j]) * np.exp(u[j] - u[jmax]) * phi[j-1]) / u[j]
#        if(jmax==1):
#            integral = 0.5 * phi[0]
        
        # Diagnostic printout.
        print("{0} {1:.3e} {2:.3e} {3:.3e}".format(j,u[j],u[j-1],u[jmax]))
        #print("{0} {1:.3e} {2:.3e} {3:.3e} {4:.3e} {5:.3e} I = {6:.3e}".format(j,u[j],u[j-1],deltauj,float(sigma_s(u[j])),u[j - jmax+1],integral))
        
    return integral

# Set a default amount of lethargy steps to 50.
def calc_phi(steps=50):
    
    # Initialize problem-specific parameters.
    Emin = 0.625           # Minimum energy of interest [eV]
    E0   = 2E6             # Maximum energy of interest [eV]
    umin = 0               # Minimum lethargy of interest.
    umax = np.log(E0/Emin) # Calculate maximum lethargy of interest.
    N_H  = 0.067           # Hydrogen number density [a/bn-cm]
    S0   = 1               # Source intensity [n/cm^2-s]
    
    # Create a linear space for all the various lethargy steps that will be 
    # iterated over.
    u = np.linspace(umin,umax,steps)

    # Create linear space to hold the resulting scalar flux values.
    phi = np.zeros(steps)

    # Determine initial (u = 0) scalar flux.
    phi[0] = S0 / sigma_t(0)
    
    for i in range(1,len(u)):
        deltaui = u[i] - u[i-1]
        print(i)
        phi[i] = (1/sigma_t(u[i])) * ((inner_integral(u,phi,i)) + S0*(sigma_s(0)/sigma_t(0))*np.exp(-u[i]))
        print("----")
        #print(i,phi[i])
        
    return u, phi

fig = plt.figure()
ax = fig.add_subplot(111)
plt.title(r"");
plt.xlabel(r"Lethargy");
plt.ylabel(r"Scalar Flux, $\phi\left(u\right)$ [n$\cdot$cm$^{-2}$\cdot$s$^{-1}$]");
ax.grid()

u, phi = calc_phi(steps=15)
plt.plot(u,phi,'ko',label="15 Steps")
#u, phi = calc_phi(steps=50)
#plt.plot(u,phi,'k^',label="50 Steps")
#u, phi = calc_phi(steps=100)
#plt.plot(u,phi,'ks',label="100 Steps")

u = np.linspace(umin,umax,100)
plt.plot(u,1/sigma_s(u),'k--',label="Analytic Sol'n.")

plt.legend(loc="best",numpoints=1)
#plt.savefig("NERS_561_HW3_JANIS_Data_Plot_Stepped.png", 
#            dpi=200, bbox_inches='tight')
```
