# Native Data Manipulation and Plotting

First, we recognize that

$f\left(\mu_{\textrm{cm}}\right)=\frac{\sum_{n=1}^{8}\frac{2n+1}{2}f_{n}P_{n}\left(\mu_{\textrm{cm}}\right)}{\sigma_{s}\left(E\right)}$

where the coefficients are given as

```python
f851tab = [1.95,0.810, 0.684, 0.512, 0.350, 0.210, 0.111, 0.061, 0.0220]
f402tab = [2.38,0.584, 0.455, 0.290, 0.147, 0.046, 0.018, 0.0043, -0.0004]
f347tab = [8.1,0.0092, 0.00015, 0, 0, 0, 0, 0, 0]
```

From here, we recognize that due to the limitations of how np.polynomial.legendre.legval() evaluates, we need to embed the $\frac{2n+1}{2}$ into to coefficients to be summed.  As such, we create new coefficient arrays that include that factor.

```python
f851tabcoeff = []
f402tabcoeff = []
f347tabcoeff = []

for n in range(1,8+1):
    f851tabcoeff.append(((2*n+1)/2)*f851tab[n])
for n,i in enumerate(f402tab):
    f402tabcoeff.append(((2*n+1)/2)*i)
for n,i in enumerate(f347tab):
    f347tabcoeff.append(((2*n+1)/2)*i)
    
print(f851tabcoeff)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.title(r"");
plt.xlabel(r"$\mu_{\textrm{cm}}$");
plt.ylabel(r"$f\left(\mu_{\textrm{cm}}\right)$");

x = np.linspace(-1, 1, 201)

#plt.ylim([-1,4])

plt.plot(x, np.polynomial.legendre.legval(x, f851tabcoeff).T/f851tab[0],'k-',label="8.51 MeV")
plt.plot(x, np.polynomial.legendre.legval(x, f402tabcoeff).T/f402tab[0],'k--',label="4.02 MeV")
plt.plot(x, np.polynomial.legendre.legval(x, f347tabcoeff).T/f347tab[0],'k:',label="34.7 keV")

plt.legend(loc=2)

#plt.show();
plt.savefig('NERS_561_HW2_Plot_1959_Data.png')
#plt.savefig('NERS_561_HW2_Plot_1959_Data.pdf',bbox_inches='tight')
```
